```python
class Library:  
    
    def __init__(biblio1, name, ammount, books):  
        biblio1.name = name
        biblio1.ammount = ammount
        biblio1.books = books
    
    def __add__(biblio1, book):
        ammount1 = biblio1.ammount + 1
        biblio1.books.append(book)
        return Library(name = biblio1.name, ammount = ammount1, books = biblio1.books)
    
    def __repr__(biblio1):
        return str(biblio1.name)
    
    def __str__(biblio1):
        return str(biblio1.name)
    
    def  __eq__(biblio1, biblio2):
        if biblio1.books == biblio2.books:
            return True
        else:
            return False
        
    def  __gt__(biblio1, biblio2):
        if biblio1.ammount > biblio2.ammount:
            return True
        else:
            return False
        
class Book:  
     
    def __init__(book1, name, author, year):  
        book1.name = name  
        book1.author = author
        book1.year = year
        
    def __repr__(book1):
        return str(book1.name)
    
    def __str__(book1):
        return str(book1.name)
 
    def  __eq__(book1, book2):        
        if book1.name == book2.name and book1.author == book2.author:
            return True
        else:
            return False
 
```
