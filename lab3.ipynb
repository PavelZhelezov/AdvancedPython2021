```python
import csv
import json
import os

vitamins=[]
for current_dir, dirs, files in os.walk("Vitamins"):
        with open('vitamins.json', 'w') as outfile:
                for file in files:
                        current_file=open(file, "r")
                        vitamin={
                                "vitamin": current_file.readline().rstrip('\n'),
                                "vitamers": current_file.readline().rstrip('\n'),
                                "solubility": current_file.readline().rstrip('\n'),
                                "daily_requirement": current_file.readline().rstrip('\n'),
                                "deficiency_diseases": current_file.readline()
                                }
                        vitamins.append(vitamin)
                json.dump(vitamins, outfile, indent=4)

jsonfile=open('vitamin.json')
print(jsonfile.read())

for current_dir, dirs, files in os.walk("Vitamins"):
        with open("vitamins.csv", "w") as vitamin.csv:
                writer = csv.writer(vitamins_csv, delimiter=",",lineterminator="\r")
                writer.writerow(["vitamins", "vitamers", "solubility", "daily_requirement", "deficiency_diseases"])
                for file in files:
                        file=open(file, "r")
                        writer.writerow([current_file.readline().rstrip('\n'), 
                                         current_file.readline().rstrip('\n'), 
                                         current_file.readline().rstrip('\n'), 
                                         current_file.readline().rstrip('\n'), 
                                         current_file.read().rstrip('\n')])

csvfile=open('vitamin.csv')
print(csvfile.read())
```
